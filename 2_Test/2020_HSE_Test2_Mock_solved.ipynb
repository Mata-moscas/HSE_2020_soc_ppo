```python
# ваше решение здесь
def caps(x):
    x = x.capitalize()     # создаем функцию: в первую очередь увеличиваем первую букву, затем разворачиваем строку
    return x[::-1]

caps('dog')
```

```python
x = input().split()       # разбиваем сплошную строку на список
for i in range(len(x)):   # поочередно применяем функцию к элементам списка
    x[i] = caps(x[i])
print(x)
```

#### 3. Задача. Факультативы (1 балл)

N студентов записываются на факультативы: немецкий, программирование, домбра, ясновидение, алгебра. Каждый студент выбирает до трех факультативов. 

Факультатив откроется, если его выбрали все студенты

Вводится число N - количество студентов. Список факультативов для каждого студента вводится отдельно с новой строки. Факультативы разделены пробелом.

**Пример ввода**  
4  
немецкий программирование  
домбра немецкий  
ясновидение алгебра  
алгебра  

**Вывод** 

0

**Пример ввода**

**Вывод**

```python
n = int(input()) # вводим число студентов
students = [] # создаем пустой список

for student in range(n): # запускаем цикл в интервале от 0 до n
    students.append(input().split()) # считываем факультативы, выбранные каждым студентом в отдельный список
```

```python
choosen = set(students[0])  # инициируем множество выбранных факультативов через первого студента, 
                            # чтобы не было пустого множества
for student in students:
    choosen &= set(student) # пересекаем наше множество с выбором каждого студента
                            # факультативы, которые были выбраны всеми - останутся в итоговом пересечении
        
print(len(choosen))
```

#### 5. Задача. Магнолия (2 балла)
*1.5 балла*  
В магазине "Магнолия" проходит акция - даются скидки на товары определенных категорий. У Маши есть список покупок, помогите ей посчитать траты на товары, которые она хочет купить. 

На ввод даются данные вида `товар - категория - количество - цена за штуку`. Каждый пункт списка вводится с новой строки. Количество пунктов может быть любым. Ввод прекращается, когда пользователь вводит 0.

Затем одной строкой даются категории, на которые сегодня скидка 15%. Категории записаны через `,`.

Посчитайте, сколько денег потратила бы Маша без скидок и сколько она заплатит в итоге. Выведите получившиеся значения на новой строке, как в примере.

*0.5 балла*  
Проверку наличия скидки реализуйте с помощью функции.

*Пример*

**Ввод**  
молоко - молочные продукты - 2 - 35  
хлеб - выпечка - 1 - 22  
творог - молочные продукты - 4 - 41  
вино - алкоголь - 1 - 453   
яблоки - фрукты - 2 - 70  
0  
алкоголь,выпечка  

**Вывод**  
849 без скидки  
777.75 со скидкой  

```python
# решение здесь
shop_list = []                              # создаем пустой список
while True:                                 # c помощью цикла while принимаем значения 
    item = input()
    if item != '0':                         # пока значения не равны нулю, принимаем их и заполняем список
        shop_list.append(item)
    else:                                   # когда вводится нуль, принимаем на ввод категории скидок и заканчиваем цикл
        categories = input().split(',')
        break

new_shop_list = []                          # создаем новый список
for i in shop_list:
    new_shop_list.append(i.split(' - '))    # наполняем его списками, полученными при разбиении строк по дефису

full_price = 0                              # создаем нулевую переменную, чтобы посчитать сумму покупки
for i in new_shop_list:                     # проходимся по каждому списку из обновленного списка и перемножаем 
    full_price += (int(i[2])*int(i[3]))     # количество на цену, на каждой итерации прибавляем это к переменной
    
def discount(purchase, category):                            # создаем функцию для расчета скидки
    if purchase[1] in category:
        price = ((int(purchase[2])*int(purchase[3])*0.85))    # если категория покупки является скидочной - то сумма ументшается
    else:                                                     # на 15%
        price = (int(purchase[2])*int(purchase[3]))           # если нет, то остается прежней   
    return price                                              # возвращается цена

discount_sum = 0                              # создаем новую переменную - сумму со скидкой
for i in new_shop_list:                       # опять проходимся по списку и применяем функцию к каждому списку с данными о товаре
    discount_sum += discount(i, categories)   # перезаписываем пременную суммы на каждой итерации

print(full_price, 'без скидки')               # вывод обоих сумм 
print(discount_sum, 'со скидкой')
```

#### 6. Задача. Проверка пароля (1.5 балла)

Каждый год студентам необходимо поменять пароль в системе LMS. C нового года к паролю стали предъявлять новые требования. 
- Он должен быть длинее 10 символов 
- В нем должна быть одна заглавная буква, но все буквы не могут быть заглавными одновременно
- В нем должна быть хотя бы одна цифра

Напишите функцию `checker()`, которая принимает на ввод гипотетический пароль и проверяет его на соответствие эти требованиям.

*Пример*

**Ввод**

checker('Taaaa1234555')

**Вывод** 

True

**Ввод**

checker('qwerty12345')

**Вывод**

False

**Ввод**

checker('taaaa')

**Вывод**

False


```python
# ваше решение здесь
def checker(x):
    if len(x) <= 10:                             # проверка на длину
        return False
    elif x.isupper() or x.islower(): # проверяем, что не все буквы заглавные, и не все строчные
        return False
    elif x.isdigit() or x.isalpha(): # проверяем, что есть и буквы и цифры, если только что-то одно, возвращаем False
        return False
    else:
        return True
    
print(checker('Taaaa1234555'))    # проверяем пароли
print(checker('qwerty12345'))
print(checker('taaaadffffffff'))
print(checker('12341224125'))
print(checker('TA2321TA22412'))
```
